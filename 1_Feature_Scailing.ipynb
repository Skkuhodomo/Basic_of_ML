```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler


a = float(input("Enter a start point: "))
b = float(input("Enter a End point:   "))

if( a < b):
  X= np.arange(a,b).astype("float32").reshape(-1,1)
  print(X.shape)


  #Simple Feature Scaling
  X_simple = X/X.max()


  # Min_max_scailing
  sc = MinMaxScaler()
  X_minmax = sc.fit_transform(X)

  # Standard Scaling (Z-score)
  sc2 = StandardScaler()
  X_standard = sc.fit_transform(X)

  #visualization
  plt.figure(figsize=(20,4))
  plt.subplot(1,4,1)
  plt.hist(X, bins=500);
  plt.title('original data')

  plt.subplot(1,4,2)
  plt.hist(X_simple, bins=500);
  plt.title('simple scaling')

  plt.subplot(1,4,3)
  plt.hist(X_standard, bins=500);
  plt.title('standard scaling')

  plt.subplot(1,4,4)
  plt.hist(X_minmax, bins=500);
  plt.title('min-max scaling');

else: #check a,b
  print("Error")
```
